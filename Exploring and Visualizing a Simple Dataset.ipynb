# Data visualizations

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. SCATTER PLOT

```python
plt.figure()
sns.scatterplot(data= file, x="sepal_length", y="petal_length")
plt.title("Scatter Plot: Sepal Length vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()
```

# 2. HISTOGRAM

```python
plt.figure()
plt.hist(file["sepal_width"], bins=10)
plt.title("Histogram: Distribution of Sepal Width")
plt.xlabel("Sepal Width")
plt.ylabel("Frequency")
plt.show()
```

# 3. BOX PLOT

```python
plt.figure()
sns.boxplot(y=file["petal_width"])
plt.title("Box Plot: Petal Width (Outliers & Spread)")
plt.ylabel("Petal Width")
plt.show()
```

Conclusion:
The Iris dataset was effectively explored and visualized using Matplotlib and Seaborn. Scatter plots identified strong correlations between numerical features, notably between sepal length and petal length. Histograms revealed the distribution and spread of sepal width, while box plots highlighted the variability and potential outliers in petal width. These visualizations transformed raw data into insightful graphical summaries, providing a solid foundation for feature selection and subsequent machine learning tasks. Such exploratory analysis is essential for informed decision-making in data science.
